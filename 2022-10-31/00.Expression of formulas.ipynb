```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
# Linear function

def linear_function(x):
    a = 0.5
    b = 2
    return a*x + b

print(linear_function(5))
```

```python
x = np.arange(-5, 5, 0.1)
y = linear_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear function')

plt.show()
```

```python
# Quadratic function

def quadratic_function(x):
    a = 1
    b = -1
    c = -2
    return a*x**2 + b*x + c

print(quadratic_function(2))
```

```python
y = quadratic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic function')

plt.show()
```

```python
# Cubic function

def cubic_function(x):
    a = 4
    b = 0
    c = -1
    d = -8
    return a*x**3 + b*x**2 + c*x + d

print(cubic_function(3))
```

```python
y = cubic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic function')

plt.show()
```

```python
# Exponential function

def exponential_function(x):
    a = 4
    return a**x

print(exponential_function(4))
```

```python
x = np.arange(-3, 2, 0.1)
y = exponential_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential function')
plt.xlim(-4, 3)
plt.ylim(-1, 15)

plt.show()
```
