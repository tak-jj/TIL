# **Preprocessing**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import os
from os.path import join
```

```python
# Prepare Data
abalone_path = join('.', 'abalone.txt')
column_path = join('.', 'abalone_attributes.txt') # '.' current directory, '..' parent directory

abalone_columns = []
for line in open(column_path):
    abalone_columns.append(line.strip())

print(abalone_columns)
```

```python
data = pd.read_csv(abalone_path, header=None, names=abalone_columns)

print(data)
```

```python
label = data.pop('Sex')
print(label)
```

```python
data.describe()
```

```python
data.info()
```

## *Scaling*

: If values differ between columns, big value have more effect to result. To prevent it, use scaling make value similary

```python
# Min-Max scaling

# Use formula
manualScaled_data = (data - np.min(data)) / (np.max(data) - np.min(data))
print(manualScaled_data)
```

```python
# Use package (recommended)
from sklearn.preprocessing import MinMaxScaler
mMscaler = MinMaxScaler()

mMscaled_data = mMscaler.fit_transform(data)
print(mMscaled_data) #return ndarray
```

```python
mMscaled_data = pd.DataFrame(mMscaled_data, columns = data.columns)

print(mMscaled_data)
```

```python
# Standard scaling
# make average 0 and str 1

# Use formula
Zdata = (data - np.mean(data)) / np.std(data)

print(Zdata)
```

```python
# Use package (recommended)
from sklearn.preprocessing import StandardScaler
sdscaler = StandardScaler()

sdscaled_data = sdscaler.fit_transform(data)
sdscaled_data = pd.DataFrame(sdscaled_data, columns = data.columns)

print(sdscaled_data)
```

## *Sampling*

If each label has different number of values, training can be hard. Sampling increase number of less value or decrease number of more value.

```python
# Random Sampling
# Pick random value and copy or delete it.
# Have risk of data bias or loss

from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

ros = RandomOverSampler()
rus = RandomUnderSampler()

oversampled_data, oversampled_label = ros.fit_resample(data, label)
undersampled_data, undersampled_label = rus.fit_resample(data, label) #return dataframe
```

```python
print('Original Data: \n{}'.format(pd.get_dummies(label).sum()))
```

```python
print('Oversampled Data: \n{}'.format(pd.get_dummies(oversampled_label).sum()))
```

```python
print('Undersampled Data: \n{}'.format(pd.get_dummies(undersampled_label).sum()))
```

```python
# SMOTE
# Oversampling method
# In less class, add value between values

from sklearn.datasets import make_classification
data, label = make_classification(
    n_samples = 1000,
    n_features = 2,
    n_redundant = 0,
    n_informative = 2,
    n_repeated = 0,
    n_classes = 3,
    n_clusters_per_class = 1,
    weights = [0.05, 0.15, 0.8],
    class_sep = 0.8,
    random_state = 2019 
) #make sample data
```

```python
fig = plt.Figure(figsize=(12,6))
plt.scatter(data[:,[0]], data[:,1], c=label, alpha=0.2)

plt.show()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()

smoted_data, smoted_label = smote.fit_resample(data, label)

print('Original Data: \n{}'.format(pd.get_dummies(label).sum()))
print('Smoted Data: \n{}'.format(pd.get_dummies(smoted_label).sum()))
```

```python
fig = plt.Figure(figsize=(12,6))
plt.scatter(smoted_data[:,[0]], smoted_data[:,1], c=smoted_label, alpha=0.2)

plt.show()
```
