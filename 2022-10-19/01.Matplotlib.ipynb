# **Matplotlib**

```python
# Import packages
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Make sample data
x = np.linspace(0, 10, 101) #np.linspace(start,end,num), make ndarray filled with evenly spaced 'num' numbers from start to end

print(x)
```

```python
# Draw graph

plt.style.use('classic') #can apply style
plt.plot(x,np.sin(x)) #plt.plot(x,y), draw graph set of (x,y)
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x)) #draw two graph in one
plt.show()
```

```python
fig = plt.figure() #make canvas

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--') #can change line style
plt.show()

fig.savefig('my_figure.png') #save figure as file
```

```python
fig.canvas.get_supported_filetypes() #can check supported file types
```

```python
# Draw multiple figure

plt.subplot(2,1,1) #.subplot(row,column,index)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))

plt.show()
```

```python
# Use axis draw graph
fig = plt.figure()
ax = plt.axes() #draw empty figure

x = np.linspace(0,10,101) 
ax.plot(x, np.sin(x)) # draw graph using axis

plt.show()
```

```python
# Style graph

# Color of line
plt.plot(x,np.sin(x),color='blue') #use color name
plt.plot(x,np.sin(x-1),color='g') #use short color code, RGB, CMYK(cyan magenta yellow black), w(white)
plt.plot(x,np.sin(x-2),color='0.5') #number between 0 to 1, 0 is black and 1 is white 
plt.plot(x,np.sin(x-3),color='#ffdd44') #Hexadecimal rgb code
plt.plot(x,np.sin(x-4),color=(1.0, 0.2, 0.3)) #other method represent rgb

plt.show()
```

```python
# Style of line
plt.plot(x, x, linestyle='solid')
plt.plot(x, x-1, linestyle='dashed')
plt.plot(x, x-2, linestyle='dashdot')
plt.plot(x, x-3, linestyle='dotted')

plt.show()
```

```python
# Other expression
plt.plot(x, x,'-')
plt.plot(x, x-1,'--')
plt.plot(x, x-2,'-.')
plt.plot(x, x-3,':')

plt.show()
```

```python
# Set color and style at once
plt.plot(x, x, '-g') 
plt.plot(x, x-1, 'c--') #can change order 
plt.plot(x, x-2, '-.k')
plt.plot(x, x-3, ':r')

plt.show()
```

```python
# Adjust axis

plt.xlim(-1,11)
plt.ylim(-5,5) #set limit of each axis

plt.plot(x, np.cos(x))
plt.show()
```

```python
# reverse

plt.xlim(11,-1) #by changing start and end can reverse figure
plt.ylim(5,-5)

plt.plot(x, np.cos(x))
plt.show()
```

```python
plt.axis([-1,11,-5,5]) #can set at once

plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight') #fit graph and figure, default

plt.show()
```

```python
plt.plot(x, np.cos(x))
plt.axis('equal') #set length of x and y axis equal

plt.show()
```

```python
# Label and legend

plt.plot(x, np.sin(x))

plt.title('f(x) = sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)') #name axis

plt.show()
```

```python
plt.plot(x, np.sin(x), 'g', label='sin(x)')
plt.plot(x, np.cos(x), 'r--', label='cons(x)') #name line

plt.legend() #show legend
plt.axis('equal')
plt.show()
```
